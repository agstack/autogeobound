# Minimal working example

```python
import mxnet as mx
from mxnet import image

import sys
sys.path.append('../../') # directory with resuneta repo

from resuneta.models.resunet_d7_causal_mtskcolor_ddist import *

# load image and reshape to valid input for network
test_img = image.imread('../data/planet/france/april/10003002.jpeg')
test_img = mx.ndarray.cast(test_img, dtype='float32')
test_img = mx.nd.moveaxis(test_img,-1,0)[:, :256, :256].expand_dims(axis=0)
test_img.shape # (1, 3, 256, 256)
```

```python
# copied from Feevos's response to GitHub issue
Nfilters_init = 32
NClasses = 6
net = ResUNet_d7(Nfilters_init, NClasses)
net.initialize()
net.hybridize()
```

```python
# this triggers the error
net(test_img)

# what I see:
# MXNetError: The size of NDArray doesn't match the requested MKLDNN memory desc. 
# MKLDNN memory requests for 262144 bytes, but got 262016 bytes from NDArray
```
