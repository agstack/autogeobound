```python
import numpy as np
import pandas as pd
import cv2
import imageio
import os
import time

import matplotlib.pyplot as plt
%matplotlib inline
```

# Downsample France Planet imagery

```python
downsize_factors = [3]

base_dir = '../data/general_blockchain/'
source_dir = os.path.join(base_dir, 'airbus_false_color', 'large', 'original')

img_names = sorted(os.listdir(source_dir))
for img_name in img_names:
    img = cv2.imread(os.path.join(source_dir, img_name))
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(cv2.cvtColor(img[:256, :256], cv2.COLOR_BGR2RGB))
    ax[0].set_title('Native Airbus resolution')
    
    for i, downsize_factor in enumerate(downsize_factors):
        new_x, new_y = img.shape[0] // downsize_factor, img.shape[1] // downsize_factor
        downsized = cv2.resize(img, (new_x, new_y))
        ax[i+1].imshow(cv2.cvtColor(downsized, cv2.COLOR_BGR2RGB))
        ax[i+1].set_title('{}x downsampled'.format(downsize_factor))
        
    plt.show()
    
    if img_name > 'airbus_geowiki_C0005':
        break
```

# Save downsampled images

```python
downsize_factors = [2] # 3

base_dir = '../data/general_blockchain/'
source_dir = os.path.join(base_dir, 'airbus_false_color', 'large', 'original')

for downsize_factor in downsize_factors:

    target_dir = os.path.join(base_dir, 'airbus_false_color', 'large', '{}x_downsample'.format(downsize_factor))
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    img_names = sorted(os.listdir(source_dir))
    for img_name in img_names:
        img = cv2.imread(os.path.join(source_dir, img_name))
        new_x, new_y = img.shape[0] // downsize_factor, img.shape[1] // downsize_factor
        downsized = cv2.resize(img, (new_x, new_y))
        filename = os.path.join(target_dir, img_name)
        cv2.imwrite(filename, downsized)
```

# Save downsampled labels

```python
base_dir = '../data/general_blockchain/airbus_labels/large/'
downsize_factors = [3] # 2

for downsize_factor in downsize_factors:
        
    source_dir = os.path.join(base_dir, 'original') # 'erosion2px'
    target_dir = os.path.join(base_dir, '{}x_downsample_original'.format(downsize_factor))
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    img_names = sorted(os.listdir(source_dir))
    already_saved = sorted(os.listdir(target_dir))
    for img_name in img_names:
        if img_name not in already_saved:
            img = cv2.imread(os.path.join(source_dir, img_name), cv2.IMREAD_UNCHANGED)
            new_x, new_y = img.shape[0] // downsize_factor, img.shape[1] // downsize_factor
            downsized = cv2.resize(img, (new_x, new_y), interpolation=cv2.INTER_NEAREST)
            filename = os.path.join(target_dir, img_name)
            cv2.imwrite(filename, downsized)
```

```python
# how many labels are completely zero?
label_dir = '../data/general_blockchain/airbus_labels/large/3x_downsample_erosion2px/'
label_names = sorted(os.listdir(label_dir))

zero_labels = []
for label_name in label_names:
    label = imageio.imread(os.path.join(label_dir, label_name))
    if np.sum(label) == 0:
        zero_labels.append(label_name)
```

```python
# erosion 1px and 2px
len(zero_labels) # 21
```

```python
# remove all but 4 of these labels from the training sample -- they can cause problems
remove_these = [int(x.split('_')[2][1:-4]) for x in zero_labels]
splits_df = pd.read_csv('../data/splits/india_planetImagery_splits_20x20_v2.csv')
new_folds = []
for i, row in splits_df.iterrows():
    if row['image_id'] in remove_these:
        new_folds.append('none')
    else:
        new_folds.append(row['fold'])
splits_df.head()
```

```python
splits_df['fold'] = new_folds
splits_df.groupby('fold').size()
```

```python
splits_df.to_csv('../data/splits/india_planetImagery_splits_20x20_v2.csv', index=False)
```
