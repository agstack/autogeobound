```python
results_dir = '../experiments/partial-france/'
models = ['fractal-resunet_3month-separate_nfilter-32_depth-6_bs-8_lr-0.001_1x-downsampled_allfields_n6759/',
          'fractal-resunet_3month-separate_nfilter-32_depth-6_bs-8_lr-0.001_2x-3x-downsampled_allfields_n6759/']
ntrain = ['1x downsampled', '2x+3x downsampled']

fig, ax = plt.subplots(1,2, figsize=(15,6))

for model in models:
    metrics = pd.read_csv(os.path.join(results_dir, model, 'metrics.csv'))
    ax[0].plot(np.arange(1, metrics.shape[0]+1), metrics['val_loss'].values)
    ax[1].plot(np.arange(1, metrics.shape[0]+1), metrics['val_mcc'].values)
    print(model)
    print(metrics['val_mcc'].max())

ax[0].set_title('Loss')
ax[0].set_ylabel('Loss')

ax[1].set_title('MCC')
ax[1].set_ylabel('MCC')

for i in range(2):
    ax[i].set_xlabel('Epoch')
plt.legend(ntrain, title='Training set size')
plt.show()
```
