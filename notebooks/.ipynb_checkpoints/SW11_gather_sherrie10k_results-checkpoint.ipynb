```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import imageio
```

# New 10k dataset results

```python
results_dir = '../experiments/france/sherrie10k/'
models = ['resunet-d6_october_class-subset_nfilter-16_bs-8_lr-0.001/',
          'resunet-d6_october_class-subset_ntrain4000_nfilter-16_bs-8_lr-0.001/',
          'resunet-d6_october_class-subset_ntrain2000_nfilter-16_bs-8_lr-0.001/']
#           'resunet-d6_october_class-subset_ntrain100_nfilter-16_bs-8_lr-0.001/']
ntrain = [6193, 4000, 2000] # 100

fig, ax = plt.subplots(1,2, figsize=(15,6))

for model in models:
    metrics = pd.read_csv(os.path.join(results_dir, model, 'metrics.csv'))
    ax[0].plot(np.arange(1, metrics.shape[0]+1), metrics['val_loss'].values)
    ax[1].plot(np.arange(1, metrics.shape[0]+1), metrics['val_mcc'].values)

ax[0].set_title('Loss')
ax[0].set_ylabel('Loss')

ax[1].set_title('MCC')
ax[1].set_ylabel('MCC')

for i in range(2):
    ax[i].set_xlabel('Epoch')
plt.legend(ntrain, title='Training set size')
plt.show()
```

```python
metrics = pd.read_csv(os.path.join(results_dir, 'resunet-d6_october_class-subset_nfilter-16_bs-8_lr-0.001/', 
                                   'metrics.csv'))
metrics['val_mcc'].values[80:].mean()
```

```python
results_dir = '../experiments/france/sherrie10k/'
models = ['resunet-d6_october_class-subset_nfilter-16_bs-8_lr-0.001/',
          'resunet-d6_october_class-subset_ntrain4000_nfilter-16_bs-8_lr-0.001/',
          'resunet-d6_october_class-subset_ntrain2000_nfilter-16_bs-8_lr-0.001/']
#           'resunet-d6_october_class-subset_ntrain100_nfilter-16_bs-8_lr-0.001/']
ntrain = [6193, 4000, 2000] # 100

fig, ax = plt.subplots(1,2, figsize=(15,6))

for model in models:
    metrics = pd.read_csv(os.path.join(results_dir, model, 'metrics.csv'))
    ax[0].plot(np.arange(1, metrics.shape[0]+1), metrics['val_f1'].values)
    ax[1].plot(np.arange(1, metrics.shape[0]+1), metrics['val_dice'].values)

ax[0].set_title('F1 score')
ax[0].set_ylabel('F1 score')

ax[1].set_title('Dice')
ax[1].set_ylabel('Dice')

for i in range(2):
    ax[i].set_xlabel('Epoch')
plt.legend(ntrain)
plt.show()
```
