# Instance segmentation post-processing

## Helper functions

```python
def InstSegm(extent, boundary, t_ext=0.4, t_bound=0.2):
    """
    INPUTS:
    extent : extent prediction
    boundary : boundary prediction
    t_ext : threshold for extent
    t_bound : threshold for boundary
    OUTPUT:
    instances
    """

    # Threshold extent mask
    ext_binary = np.uint8(extent >= t_ext)

    # Artificially create strong boundaries for
    # pixels with non-field labels
    input_hws = np.copy(boundary)
    input_hws[ext_binary == 0] = 1

    # Create the directed graph
    size = input_hws.shape[:2]
    graph = hg.get_8_adjacency_graph(size)
    edge_weights = hg.weight_graph(
        graph,
        input_hws,
        hg.WeightFunction.mean
    )

    tree, altitudes = hg.watershed_hierarchy_by_dynamics(
        graph,
        edge_weights
    )
    
    # Get individual fields
    # by cutting the graph using altitude
    instances = hg.labelisation_horizontal_cut_from_threshold(
        tree,
        altitudes,
        threshold=t_bound)
    
    instances[ext_binary == 0] = -1

    return instances
```

## IoU

```python
def get_IoUs(extent_true, extent_pred, boundary_pred, t_ext=0.4, 
             t_bound=0.2, plot=False):
    
    # get predicted instance segmentation
    instances_pred = InstSegm(extent_pred, boundary_pred, t_ext=t_ext, t_bound=t_bound)
    instances_pred = measure.label(instances_pred, background=-1)
    
    # get instances from ground truth label
    binary_true = extent_true > 0
    instances_true = measure.label(binary_true, background=0)
    
    # remove non-cropland
    # kernel = np.ones((2,2), np.uint8)
    # cropland_mask = si.binary_closing(binary_true, structure=kernel, 
    #                                   iterations=4, border_value=1)
    # instances_pred = instances_pred * cropland_mask
    
    if plot:
        fig, ax = plt.subplots(1, 2, figsize=(8, 4))
        ax[0].imshow(instances_true)
        ax[1].imshow(instances_pred)
        plt.show()
    
    # loop through true fields
    field_values = np.unique(instances_true)
    best_IoUs = []
    field_sizes = []
    
    for field_value in field_values:
        if field_value == 0:
            continue # move on to next value
            
        this_field = instances_true == field_value
        field_sizes.append(np.sum(this_field))
        
        # find predicted fields that intersect with true field
        intersecting_fields = this_field * instances_pred
        intersect_values = np.unique(intersecting_fields)
        
        intersect_fields = np.isin(instances_pred, intersect_values[1:])
        
        if plot:
            fig, ax = plt.subplots(1, 2, figsize=(8, 4))
            ax[0].imshow(this_field)
            ax[1].imshow(intersect_fields)
            plt.show()
        
        # compute IoU for each intersecting field 
        field_IoUs = []
        for intersect_value in intersect_values:
            if intersect_value == 0:
                continue # move on to next value
            pred_field = instances_pred == intersect_value
            union = this_field + pred_field > 0
            intersection = (this_field * pred_field) > 0
            IoU = np.sum(intersection) / np.sum(union)
            field_IoUs.append(IoU)
    
        # take maximum IoU - this is the IoU for this true field
        if len(field_IoUs) > 0:
            best_IoUs.append(np.max(field_IoUs))
        else:
            best_IoUs.append(0)
    
    return best_IoUs, field_sizes
```

```python
# hyperparameter values
t_exts = np.linspace(0.3, 0.7, 9)
t_bounds = np.linspace(0.1, 0.5, 9)

# run for all test fields
downsamples = ['2x', '3x']

label_dir = '../data/planet/france/sherrie10k/extent_labels/1250px/'
model_name = 'fractal-resunet_3month-separate_nfilter-32_depth-6_bs-8_lr-0.001_2x-3x-downsampled_allfields_n6759'
results_dir = '../results/partial-france/{}/consensus_AprJulOct'.format(model_name)

pred_names = os.listdir(os.path.join(results_dir, 'extent_predictions'))
image_ids = []
for image_id in val_ids:
    add = True
    for downsample in downsamples:
        if image_id + '-{}.npy'.format(downsample) not in pred_names:
            add = False
    if add:
        image_ids.append(image_id)
image_ids = image_ids[:200]

mIoUs = []
IoU_50s = []
for t_ext in t_exts:
    for t_bound in t_bounds:
        IoUs = []

        for i, image_id in enumerate(image_ids):
            if i % 100 == 0:
                print(i)
            for scale in downsamples:
                label_path = os.path.join(
                    label_dir, '{}_downsample/'.format(scale), '{}.png'.format(image_id))
                if os.path.exists(label_path):
                    example_truth = imageio.imread(label_path)
                    extent_pred = np.load(os.path.join(results_dir, 'extent_predictions/{}-{}.npy'.format(image_id, scale)))
                    boundary_pred = np.load(os.path.join(results_dir, 'boundary_predictions/{}-{}.npy'.format(image_id, scale)))
                    img_IoUs, _ = get_IoUs(
                        example_truth, extent_pred, boundary_pred, t_ext=t_ext, t_bound=t_bound)
                    IoUs = IoUs + img_IoUs
                    
        mIoUs.append(np.mean(IoUs))
        IoU_50s.append(np.sum(np.array(IoUs) > 0.5) / len(IoUs))

hp_df = pd.DataFrame({
    't_ext': np.repeat(t_exts, len(t_bounds)),
    't_bound': np.tile(t_bounds, len(t_exts)),
    'mIoU': mIoUs,
    'IoU_50': IoU_50s
})
```

```python
splits_path = '../data/splits/sherrie10k_planetImagery_splits_20x20_4x-downsampled.csv'
splits_df = pd.read_csv(splits_path, dtype=str)

val_ids = splits_df[splits_df['fold'] == 'val']['image_id'].values
month = '2019_10'
data_dir = '../data/planet/france/sherrie10k/'
model_name = 'resunet-d6_2019_10_nfilter-16_bs-8_lr-0.001_4x-downsampled_allfield_n6759'
results_dir = '../results/france/sherrie10k/{}/{}'.format(model_name, month)

truth_sizes = {x: [] for x in val_ids}
pred_sizes = {x: [] for x in val_ids}
truth_hit = {x: [] for x in val_ids}
pred_hit = {x: [] for x in val_ids}
truth_iou = {x: [] for x in val_ids}
pred_iou = {x: [] for x in val_ids}

for j, image_id in enumerate(val_ids):
    if j % 10 == 0:
        print(j)

    extent_pred_path = os.path.join(results_dir, 'extent_predictions/{}_{}.npy'.format(image_id, month))
    if os.path.exists(extent_pred_path):
        example_img = imageio.imread(
            os.path.join(data_dir, 
                         'monthly_mosaics_renamed_clipped_merged/1250px/4x_downsample/{}/{}_{}.tif'.format(month, image_id, month)))
        example_truth = imageio.imread(os.path.join(data_dir, 'extent_labels/1250px/4x_downsample/{}.png'.format(image_id)))
        example_extent = np.load(extent_pred_path)
        example_boundary = np.load(os.path.join(results_dir, 'boundary_predictions/{}_{}.npy'.format(image_id, month)))

        instances = InstSegm(example_extent, example_boundary)

        # map instances to better range for plotting
        unique_instances = sorted(np.unique(instances))
        n_instances = len(unique_instances)
        instance_map = {x: i for i, x in enumerate(unique_instances)}
        def map_values(x):
            return instance_map[x]
        instances_mapped = np.array(list(map(map_values, instances.flatten()))).reshape( 
            instances.shape[0], instances.shape[1])

        # mask out non-crop areas using ground truth
        # instances_mapped * (example_truth[:256,:256] > 0)
        # grow ground truth by 2 pixels everywhere
        kernel = np.ones((2,2), np.uint8)
        binary_truth = example_truth > 0
        instances_truth = measure.label(binary_truth, background=0)
        cropland_mask = si.binary_closing(binary_truth, structure=kernel, iterations=4, border_value=1)
        instances_masked = instances_mapped * cropland_mask

        for i in range(1, len(np.unique(instances_truth))):
            truth_sizes[image_id].append(np.sum(instances_truth == i))
        for i in np.unique(instances_masked)[1:]:
            pred_sizes[image_id].append(np.sum(instances_masked == i))

        # hit rate of ground truth fields
        for i in range(1, len(np.unique(instances_truth))):
            truth_field = instances_truth == i
            truth_field_size = np.sum(truth_field)
            pred_overlap = truth_field * instances_masked
            pred_overlap_instances = np.unique(pred_overlap, return_counts=True)[0]
            pred_overlap_sizes = np.unique(pred_overlap, return_counts=True)[1]
            if len(pred_overlap_sizes) < 2:
                truth_hit[image_id].append(0)
            else:
                max_overlap_size = np.max(pred_overlap_sizes[1:])
                truth_hit[image_id].append(max_overlap_size / truth_field_size)
                pred_field_id = pred_overlap_instances[1:][np.argmax(pred_overlap_sizes[1:])]
                pred_field_size = np.sum(instances_masked == pred_field_id)
                union_size = truth_field_size + pred_field_size - max_overlap_size
                truth_iou[image_id].append(max_overlap_size / union_size)

        # hit rate of predicted parcels
        for i in np.unique(instances_masked)[1:]:
            pred_field = instances_masked == i
            pred_field_size = np.sum(pred_field)
            truth_overlap = pred_field * instances_truth
            truth_overlap_instances = np.unique(truth_overlap, return_counts=True)[0]
            truth_overlap_sizes = np.unique(truth_overlap, return_counts=True)[1]
            if len(truth_overlap_sizes) < 2:
                pred_hit[image_id].append(0)
            else:
                max_overlap_size = np.max(truth_overlap_sizes[1:])
                pred_hit[image_id].append(max_overlap_size / pred_field_size)
                truth_field_id = truth_overlap_instances[1:][np.argmax(truth_overlap_sizes[1:])]
                truth_field_size = np.sum(instances_truth == truth_field_id)
                union_size = pred_field_size + truth_field_size - max_overlap_size
                pred_iou[image_id].append(max_overlap_size / union_size)
```

```python
# only field crops
field_crops = np.array(list(np.arange(1,10)) + [14,15,16])

splits_path = '../data/splits/sherrie10k_planetImagery_splits_20x20_4x-downsampled.csv'
splits_df = pd.read_csv(splits_path, dtype=str)

val_ids = splits_df[splits_df['fold'] == 'val']['image_id'].values
month = '2019_10'
data_dir = '../data/planet/france/sherrie10k/'
model_name = 'resunet-d6_2019_10_nfilter-16_bs-8_lr-0.001_4x-downsampled_allfield_n6759'
results_dir = '../results/france/sherrie10k/{}/{}'.format(model_name, month)

truth_sizes_fieldcrops = {x: [] for x in val_ids}
truth_hit_fieldcrops = {x: [] for x in val_ids}

for j, image_id in enumerate(val_ids):
    if j % 10 == 0:
        print(j)

    extent_pred_path = os.path.join(results_dir, 'extent_predictions/{}_{}.npy'.format(image_id, month))
    if os.path.exists(extent_pred_path):
        example_img = imageio.imread(
            os.path.join(data_dir, 
                         'monthly_mosaics_renamed_clipped_merged/1250px/4x_downsample/{}/{}_{}.tif'.format(month, image_id, month)))
        example_truth = imageio.imread(os.path.join(data_dir, 'extent_labels/1250px/4x_downsample/{}.png'.format(image_id)))
        example_extent = np.load(extent_pred_path)
        example_boundary = np.load(os.path.join(results_dir, 'boundary_predictions/{}_{}.npy'.format(image_id, month)))

        instances = InstSegm(example_extent, example_boundary)

        # map instances to better range for plotting
        unique_instances = sorted(np.unique(instances))
        n_instances = len(unique_instances)
        instance_map = {x: i for i, x in enumerate(unique_instances)}
        def map_values(x):
            return instance_map[x]
        instances_mapped = np.array(list(map(map_values, instances.flatten()))).reshape( 
            instances.shape[0], instances.shape[1])

        # mask out non-crop areas using ground truth
        # instances_mapped * (example_truth[:256,:256] > 0)
        # grow ground truth by 2 pixels everywhere
        kernel = np.ones((2,2), np.uint8)
        example_truth = np.isin(example_truth, field_crops) # mask out non-field crops
        binary_truth = example_truth > 0
        instances_truth = measure.label(binary_truth, background=0)
        cropland_mask = si.binary_closing(binary_truth, structure=kernel, iterations=4, border_value=1)
        instances_masked = instances_mapped * cropland_mask

        for i in range(1, len(np.unique(instances_truth))):
            truth_sizes_fieldcrops[image_id].append(np.sum(instances_truth == i))
        
        # hit rate of ground truth fields
        for i in range(1, len(np.unique(instances_truth))):
            truth_field = instances_truth == i
            truth_field_size = np.sum(truth_field)
            pred_overlap = truth_field * instances_masked
            pred_overlap_sizes = np.unique(pred_overlap, return_counts=True)[1]
            if len(pred_overlap_sizes) < 2:
                truth_hit_fieldcrops[image_id].append(0)
            else:
                max_overlap_size = np.max(pred_overlap_sizes[1:])
                truth_hit_fieldcrops[image_id].append(max_overlap_size / truth_field_size)
```

```python
splits_path = '../data/splits/sherrie10k_planetImagery_splits_20x20_4x-downsampled.csv'
splits_df = pd.read_csv(splits_path, dtype=str)

val_ids = splits_df[splits_df['fold'] == 'val']['image_id'].values
months = ['2019_04', '2019_07', '2019_10']
data_dir = '../data/planet/france/sherrie10k/'
model_name = 'resunet-d6_2019_10_nfilter-16_bs-8_lr-0.001_4x-downsampled_allfield_n6759'

truth_sizes = {x: [] for x in val_ids}
pred_sizes = {x: [] for x in val_ids}
truth_hit = {x: [] for x in val_ids}
pred_hit = {x: [] for x in val_ids}

for j, image_id in enumerate(val_ids):
    if j % 10 == 0:
        print(j)
        
    example_truth = imageio.imread(os.path.join(data_dir, 'extent_labels/1250px/4x_downsample/{}.png'.format(image_id)))

    if os.path.exists(os.path.join('../results/france/sherrie10k', model_name, '2019_10', 'extent_predictions', '{}_2019_10.npy'.format(image_id))) \
        and os.path.exists(os.path.join('../results/france/sherrie10k', model_name, '2019_07', 'extent_predictions', '{}_2019_07.npy'.format(image_id))) \
        and os.path.exists(os.path.join('../results/france/sherrie10k', model_name, '2019_04', 'extent_predictions', '{}_2019_04.npy'.format(image_id))):
        
        all_extent = np.zeros((len(months), 256, 256))
        all_boundary = np.zeros((len(months), 256, 256))
        
        for m, month in enumerate(months):
            results_dir = '../results/france/sherrie10k/{}/{}'.format(model_name, month)
            extent_pred_path = os.path.join(results_dir, 'extent_predictions/{}_{}.npy'.format(image_id, month))
            
            all_extent[m] = np.load(extent_pred_path)
            all_boundary[m] = np.load(os.path.join(results_dir, 'boundary_predictions/{}_{}.npy'.format(image_id, month)))
            
        example_extent = all_extent.mean(axis=0)
        example_boundary = all_boundary.mean(axis=0)
        instances = InstSegm(example_extent, example_boundary)

        # map instances to better range for plotting
        unique_instances = sorted(np.unique(instances))
        n_instances = len(unique_instances)
        instance_map = {x: i for i, x in enumerate(unique_instances)}
        def map_values(x):
            return instance_map[x]
        instances_mapped = np.array(list(map(map_values, instances.flatten()))).reshape( 
            instances.shape[0], instances.shape[1])

        # mask out non-crop areas using ground truth
        # instances_mapped * (example_truth[:256,:256] > 0)
        # grow ground truth by 2 pixels everywhere
        kernel = np.ones((2,2), np.uint8)
        binary_truth = example_truth > 0
        instances_truth = measure.label(binary_truth, background=0)
        cropland_mask = si.binary_closing(binary_truth, structure=kernel, iterations=4, border_value=1)
        instances_masked = instances_mapped * cropland_mask

        for i in range(1, len(np.unique(instances_truth))):
            truth_sizes[image_id].append(np.sum(instances_truth == i))
        for i in np.unique(instances_masked)[1:]:
            pred_sizes[image_id].append(np.sum(instances_masked == i))

        # hit rate of ground truth fields
        for i in range(1, len(np.unique(instances_truth))):
            truth_field = instances_truth == i
            truth_field_size = np.sum(truth_field)
            pred_overlap = truth_field * instances_masked
            pred_overlap_sizes = np.unique(pred_overlap, return_counts=True)[1]
            if len(pred_overlap_sizes) < 2:
                truth_hit[image_id].append(0)
            else:
                max_overlap_size = np.max(pred_overlap_sizes[1:])
                truth_hit[image_id].append(max_overlap_size / truth_field_size)

        # hit rate of predicted parcels
        for i in np.unique(instances_masked)[1:]:
            pred_field = instances_masked == i
            pred_field_size = np.sum(pred_field)
            truth_overlap = pred_field * instances_truth
            truth_overlap_sizes = np.unique(truth_overlap, return_counts=True)[1]
            if len(truth_overlap_sizes) < 2:
                pred_hit[image_id].append(0)
            else:
                max_overlap_size = np.max(truth_overlap_sizes[1:])
                pred_hit[image_id].append(max_overlap_size / pred_field_size)
```
