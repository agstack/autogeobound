# Radiant Earth Foundation Field Boundaries

## Kenya

```python
base_dir = '../../mlhub-tutorials/data/ref_african_crops_kenya_01_labels/'
label_dirs = [x for x in os.listdir(base_dir) if 'ref' in x]

dataframes_list = []
for label_dir in label_dirs:
    label_path = os.path.join(base_dir, label_dir, 'labels.geojson')
    print(gpd.read_file(label_path).shape)
    dataframes_list.append(gpd.read_file(label_path))
labels = gpd.GeoDataFrame(pd.concat(dataframes_list, ignore_index=True), crs=dataframes_list[0].crs)
labels.to_file("../data/parcels/radiant/ref_african_crops_kenya_01_labels.shp")
print(labels.shape)
```
